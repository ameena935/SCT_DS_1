```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data_2024={
    'Country':['India','China','United States','Indonesia','Pakistan','Nigeria','Brazil','Bangladesh','Russia','Ethiopia'],
    'Population (Millions)':[1441.72,1425.18,341.81,279.80,245.21,229.15,217.64,174.70,143.96,129.72]
}
df = pd.DataFrame(data_2024)

plt.figure(figsize=(12,8))
sns.set_style("whitegrid")

barplot=sns.barplot(x='Population (Millions)',y='Country',data=df,hue='Country',palette='viridis')

plt.title('Top 10 Countries by Total Population in 2024',fontsize=16,pad=20)
plt.xlabel('Population (Millions)',fontsize=14)
plt.ylabel('Country',fontsize=14)

for i, v in enumerate(df['Population (Millions)']):
    barplot.text(v+10,i,f'{v:,.0f}M',va='center',fontsize=10,color='black')

plt.tight_layout()
plt.show()
```

```python
data = {
    'Age Group':['0-14 Years\n(Children)','15-64 Years\n(Working Age)','65+ Years\n(Elderly)'],
    'Percentage':[25,65,10],
    'Approx_Population_Billions':[2.0,5.3,0.8] 
}
df = pd.DataFrame(data)

plt.figure(figsize=(10,6))
sns.set_style("whitegrid")

custom_colors=['#66b3ff','#99ff99','#ffcc99']

barplot=sns.barplot(x='Age Group',y='Percentage',data=df,hue='Age Group',palette=custom_colors)

plt.title('Global Population Distribution by Age Group in 2024',fontsize=16,pad=20)
plt.ylabel('Percentage of Total Population (%)',fontsize=14)
plt.xlabel('Age Category',fontsize=14)
plt.ylim(0,80)  

for i, v in enumerate(df['Percentage']):
    barplot.text(i,v + 1,f'{v}%',ha='center',fontsize=14,fontweight='bold')
    
    pop = df['Approx_Population_Billions'][i]
    barplot.text(i,v - 5,f'~{pop} Billion',ha='center',fontsize=11,color='black')

plt.tight_layout()
plt.show()
```

```python
data = {
    'Age Group':['0-14 Years\n(Children)','15-64 Years\n(Working Age)','65+ Years\n(Elderly)'],
    'Percentage':[25,68,7],
    'Population_Billions':[0.36,0.98,0.10]
    }
df = pd.DataFrame(data)

plt.figure(figsize=(10,6))
sns.set_style("whitegrid")

age_colors=['#ff9999',"#73AADE",'#99ff99']

barplot=sns.barplot(x='Age Group',y='Percentage',data=df,hue='Age Group',palette=age_colors)

plt.title('Population Distribution by Age Group in India (2024)',fontsize=16,pad=20)
plt.ylabel('Percentage of Population (%)',fontsize=14)
plt.xlabel('Age Category',fontsize=14)
plt.ylim(0,80) 

for i, row in df.iterrows():
    barplot.text(i, row['Percentage']+1, 
                 f"{row['Percentage']}%", 
                 ha='center',fontsize=14,fontweight='bold',color='black')
    
    barplot.text(i, row['Percentage']-5, 
                 f"~{row['Population_Billions']} Billion", 
                 ha='center',fontsize=11,color='black')

plt.tight_layout()
plt.show()
```

```python
data = {
    'Gender':['Male','Female'],
    'Population_Billions':[4.10,4.06]
}
df = pd.DataFrame(data)

total_pop=df['Population_Billions'].sum()
df['Percentage']=(df['Population_Billions']/total_pop)*100

plt.figure(figsize=(8,6))
sns.set_style("whitegrid")

gender_colors=['#1f77b4','#e377c2'] 

barplot=sns.barplot(x='Gender',y='Population_Billions',data=df,hue='Gender',palette=gender_colors)

plt.title('Global Population Distribution by Gender (2024)',fontsize=16,pad=20)
plt.ylabel('Population (Billions)',fontsize=14)
plt.xlabel('Gender',fontsize=14)
plt.ylim(0,5)  

for i,row in df.iterrows():
    barplot.text(i,row['Population_Billions']+0.05, 
                 f"{row['Population_Billions']}B", 
                 ha='center',fontsize=14,fontweight='bold',color='black')
    
    barplot.text(i, row['Population_Billions']-0.5, 
                 f"({row['Percentage']:.1f}%)", 
                 ha='center',fontsize=13,color='white',fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
data={
    'Country':['Vatican City', 'Tuvalu', 'Nauru', 'Palau', 'San Marino','Marshall Islands', 'Monaco', 'Liechtenstein', 'Saint Kitts & Nevis', 'Dominica'],
    'Population': [764, 11478, 12884, 18051, 33581, 
                   37548, 38631, 39870, 46843, 73040]
}
df=pd.DataFrame(data)
df=df.sort_values('Population',ascending=True)

plt.figure(figsize=(12,8))
sns.set_style("whitegrid")

barplot=sns.barplot(x='Population',y='Country',data=df,hue='Country',palette='viridis')

plt.title('Top 10 Least Populous Sovereign Countries (2024)',fontsize=16,pad=20)
plt.xlabel('Population',fontsize=14)
plt.ylabel('Country',fontsize=14)

for i,v in enumerate(df['Population']):
    barplot.text(v + 1000,i,f'{v:,.0f}',va='center',fontsize=11,fontweight='bold',color='black')

plt.tight_layout()
plt.show()
```
